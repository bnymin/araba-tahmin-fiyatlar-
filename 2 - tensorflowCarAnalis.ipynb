```python
dataFrame=dataFrame[dataFrame.year != 1970]
```

```python
dataFrame.groupby("year").mean()["price"]
```

```python
dataFrame.drop("transmission" ,axis =1,inplace=True)
```

```python
y = dataFrame["price"].values
x = dataFrame.drop("price", axis = 1).values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train, y_test = train_test_split(x,y,test_size = 0.33,random_state = 10)
```

```python
len(x_train)
```

```python
len(x_test)
```

üstekki satırlarda veriyi test train olarak ikiye bölüp hazırlık yaptık
allata ise model oluşturup veri yapay zeka ile tahminini sağladık

```python

from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(11 ,activation = "relu"))
model.add(Dense(11 ,activation = "relu"))
model.add(Dense(11 ,activation = "relu"))
model.add(Dense(11 ,activation = "relu"))
model.add(Dense(11 ,activation = "relu"))


model.add(Dense(1))

model.compile(optimizer="adam" , loss ="mse")
```

```python
model.fit(x=x_train,y=y_train , validation_data=(x_test,y_test),batch_size=300,
          epochs=350)
```

```python
kayıpVeri = pd.DataFrame(model.history.history)
```

```python
kayıpVeri
```

```python
kayıpVeri.plot()
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
tahminDizizsi =model.predict(x_test)
```

```python
tahminDizizsi
```

```python
mean_absolute_error(y_test,tahminDizizsi)# tahmin dizimizin sapma payı
```

```python
mean_squared_error(y_test,tahminDizizsi)
```
